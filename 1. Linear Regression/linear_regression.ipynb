## 2.3. Regularization


In this section, we will find the coefficients that minimize the error for a $M=9$ degree polynimal given the regularization coefficient $\lambda$, over $\lambda \in \{0, 10^{-8}, 10^{-7}, 10^{-6}, \cdots, 10^{-1}, 10^{0} (=1) \}$.

```python
fig, ax = plt.subplots()

lambdas = np.array([0] + [pow(10, x) for x in [-8, -7, -6, -5, -4, -3, -2, -1, 0]])

w_list = [linear_regression.closed_form(X_train, y_train, lam = l) for l in lambdas]

loss_tr = [1/(X_train.shape[0])*sum([(np.matmul(X_train[i], w.T) - y_train[i])**2 for i in range(X_train.shape[0])]) for w in w_list]
loss_te = [1/(X_test.shape[0])*sum([(np.matmul(X_test[i], w.T) - y_test[i])**2 for i in range(X_test.shape[0])]) for w in w_list]

train_rmse = [np.sqrt(1/(X_train.shape[0])*ltr) for ltr in loss_tr] #np.sqrt(1/m*np.sum((y_train-np.matmul(x_train[:,:n+1], w[m]))**2) for n in M_candidates) #np.sqrt(1/m*([sum(np.dot(w[i].T, x_train[i])- y_train[i])**2 for i in M_candidates]))
test_rmse = [np.sqrt(1/(X_test.shape[0])*lte) for lte in loss_te] #np.sqrt(1/m*np.sum((y_test-np.matmul(x_test[:,:n+1], w[m]))**2) for n in M_candidates) #np.sqrt(1/m*([sum(np.dot(w[i].T, x_test[i])- y_test[i])**2 for i in M_candidates]))

######################## Generating a Plot ########################
ax.grid()
ax.set(xlabel=r"$\log_{10} \lambda$", ylabel="RMS Error")

log_lambdas = np.log10(lambdas + np.array([1e-9, *np.zeros(9)]))
log_lambda_labels = ["$-\infty$", *log_lambdas[1:]]
ax.set_xticks(log_lambdas, labels=log_lambda_labels)
ax.plot(log_lambdas, train_rmse, linewidth = 2.0, label = 'Train')
ax.plot(log_lambdas, test_rmse, linewidth = 2.0, label = 'Test')
plt.legend()
plt.show()
```
